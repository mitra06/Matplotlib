```python
# import required libraries
import matplotlib.pyplot as plt
import numpy as np
```

```python
# set the value of x and y axis
x=np.arange(1,11)
y=np.arange(2,22,2)
```

```python
#plot the x and y values
plt.plot(x,y)
#name of x axis
plt.xlabel("X axis")
# name of y axis
plt.ylabel("Y axis")
# title of my graph
plt.title("My first graph")
#function to show the graph
plt.show()
```

```python
#plot two or more lines on same plot
x1=[4,1,4,10,8]
y1=[6,3,2,5,9]
plt.plot(x,y, label="Line1")
plt.plot(x1,y1, label="Line2")
plt.xlabel("x axis")
plt.ylabel("Y axis")
plt.legend()
plt.show()
```

```python
#some customizations
plt.plot(x,y,color="red",linestyle="dashed",marker="o",markerfacecolor="blue",markersize=12,linewidth=4)
plt.xlim(1,10)
plt.ylim(1,20,2)
plt.xlabel("x-axis")
plt.xlabel("y-axis")
plt.title("customized plotting")
plt.show()
```

```python
#Bar chart

age=[3,5,8,12,14]
height=[3,4,4,5,6]
plt.bar(age,height,align="center",color=['red','green'],width=1.0)
plt.xlabel("age")
plt.ylabel("height")
plt.plot()
plt.show()
```

```python
ages = [2,5,70,40,30,45,50,45,43,40,44,60,7,13,57,18,90,77,32,21,20,40]
range=(0,100)
bins=10
plt.hist(ages,bins,range,histtype='bar',color='red',rwidth = 0.8)
plt.xlabel("age")
plt.ylabel("no of peoples")
plt.show()
```

```python
plt.scatter(x,y,label="star",color='r',marker='*',s=40)
plt.show()
```
