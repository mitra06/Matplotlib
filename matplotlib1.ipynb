```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# for example i have an equations 2x+5
x=np.arange(1,11)
y=2*x+5
plt.title("My first tutorial on matplotlib")
plt.xlabel("X label data")
plt.ylabel("Y label data")
plt.plot(x,y)
plt.show()
```

```python
# plot with some attributes
plt.plot(x,y,"bo")
plt.show()
```

```python
plt.plot(x,y,"r-")
plt.show()
```

```python
plt.plot(x,y,"r-.")
plt.show()
```

```python
# sine wave plot
x = np.arange(0,3 * np.pi, 0.1)
y=np.sin(x)
plt.title("sine wave form")
plt.plot(x,y)
plt.show()
```

```python
x=np.arange(0,3*np.pi,0.1)
y=np.cos(x)
plt.title("cos wave form")
plt.plot(x,y)
plt.show()
```

```python
x=np.arange(0,3*np.pi,0.1)
y=np.tan(x)
plt.title("tan wave form")
plt.plot(x,y)
plt.show()
```

```python
#Subplot
#The subplot() function allow you to plot different things in same figure

x=np.arange(0,3*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)
y_tan=np.tan(x)
#setup subplot grid items which height will 2 and width 1
# and set first subplot as active

plt.subplot(3,1,1)
# make the first plot
plt.plot(x,y_sin)
plt.title("sine")
plt.subplot(3,1,2)
plt.plot(x,y_cos)
plt.title("cosine")
plt.subplot(3,1,3)
plt.plot(x,y_tan)
plt.title("tan")
plt.show()
```

```python
#Bar plot
x=[5,8,10,12]
y=[12,14,16,10]
plt.plot(x,y)
plt.show()
```

```python
plt.bar(x,y)
plt.show()
```

```python
plt.bar(x,y,align="center", color="r")
plt.show()
```

```python
# histogram generates graphical representation of the frequency distribution of the data
a=np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
plt.hist(a,bins=[0,20,40,60,80,100])
plt.title("histogram")
plt.show()
```

```python
age=np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
plt.hist(age,bins=20)
plt.title("Age distributions using histogram")
plt.xlabel("Age")
plt.ylabel("person")
plt.show()
```

```python
#boxplot
plt.boxplot(age)
plt.show()
```
